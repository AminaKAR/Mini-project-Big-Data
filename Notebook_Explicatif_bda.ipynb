## Les fonctions définissant les différentes catégories

### Les tortues régulières

Les tortues  régulières sont connu par leur vitesse toujours constante. On a choisit de calculer la standard deviation. Si on la trouve égale à zéro on est bien dans le cas de tortue réguliére.

On cite qu'il ya des cas ou on va trouver des valeurs aberrantes pour la standard deviation, comme le cas de la tortue 6 tiny ou on trouve une valeur de 690 alors que toutes les autres valeurs sont contantes et égales à 345.

```python
def regular(tab):
    std=pd.core.series.Series(tab).std()
    if std == 0:
        return(True,'Tortue régulière avec une vitesse : '+str(tab[0]))
    else:
        return(False,'0')
```

### Les tortues distraites

Les tortues distraites sont les tortues qui avancent avec une vitesse ( nombre de pas aléatoire ) entre une valeur minimale et une autre maximale. 

On peut dire que tout tortue qui n'est pas réguliére, pas cyclique ni fatiguée elle est jugée comme distraite.

```python
def distraite(tab):
    tab=pd.core.series.Series(tab)
    return(True,'Tortue distraite avec un max : '+str(tab.max())+' et un min : '+str(tab.min()))
```

Finalement si on arrive à la fin avec un rappor entre le nombre de zéro ( les vhangement de vitesses égaux) et la longeur de la liste >= 0.8  on juge que la tortue est bien fatiguée

```python
def tired(speed):
    s=list(speed)
    if 0 not in s :
        return(False,'0')
    size = len(list(s)) 
    idx_list = [idx + 1 for idx, val in
                enumerate(s) if val == 0] 


    res = [s[i: j] for i, j in
            zip([0] + idx_list, idx_list + 
            ([size] if idx_list[-1] != size else []))] 


    res=(res[1:-1])
    nbrzero=0
    for i in res:
        tmp=pd.Series(i[:]).diff().diff().dropna()#we eliminate the nan values
        if len(list(tmp))-list(tmp).count(0)==3:
            nbrzero+=1

        if (nbrzero/(len(tmp)+1)) >0.8:
            return (True,'fatigué')
    return(False,'0')
```

### Les tortues cycliques 

Maintenant les tortues cycliques, eux elles  avancent avec une vitesse différente
dans une méme fenétre qui va se répéter aprés.

Comme nous avons besoin d'au moins trois points pour faire deux segments, on fait le test sur le nombre de top pour éviter les cas ou la tortue n'est pas cyclique ou on n'as pas assez de données donc aucun moyen d'être sûr.

Aprés nous vérifions si les indices sont régulièrement espacés, c'est une heuristique si l'élément est unique dans le cycle, c'est léger en calcul et vaut la peine d'essayer.

Maintenant si c'est Vrai alors on peut directement dire que c'est cyclique, cependant si c'est faux on ne peut pas être sûr que ce n'est pas cyclique ou si l'hypothèse de l'unicité de l'élément est fausse.


```python
def get_indices(elt,list_of_elts):
    import more_itertools as iter_tools 
    indices = list(iter_tools.locate(list_of_elts, lambda element: element == elt))
    return(indices)
```

```python
def cyclic(speeds): 
    """ 
    a cycle has no begining and no end hence the speeds of the 
    cycle window won't be sorted like the server configuration
    we check the indices of an element , we assume its unicity in the cycle to avoid heavy computing
    if it's not the case we use an exhaustive search
    """
    indices=get_indices(speeds[0],speeds)
    if len(indices)<3:
        return False,'0' # we need at least three points to make two segments, this is due if the turtle is not cyclic or not enough data and no way to be  sure

    null_indices_std=pd.Series(indices).diff()[1:].std()==0 # we verify if the indices are evenly spaced, it's a heuristic if the elt is unique in the cycle, it's computationaly light and worth trying
    if null_indices_std :
        cycle=True
        for i in range(len(indices)-2):
            if speeds[indices[i]:indices[i+1]]!=speeds[indices[i+1]:indices[i+2]]:
                cycle=False
                break
        if cycle :
            return(True,'Tortue cyclique, le tableau des vitesses dans la fenêtre est : '+str(speeds[indices[0]:indices[1]]))
        #if it's True then it's cyclic, however if it's false we c'ant be sure if it's not cyclic or if the assumption of the element's unicity is wrong
      
    potential_cycles=[]
    for i in range (len(indices)):
        for j in range(i+1,len(indices)):
            #print('*'*5)
            #print(indices[i],indices[j],abs(indices[i]-indices[j]))
            long_seq_to_check=abs(indices[i]-indices[j])
            begining=indices[j]
            end=indices[j]+long_seq_to_check
            cycle=True
            while(end<len(speeds)):
                if(speeds[indices[i]:indices[j]]==speeds[begining:end]):
                    """print(True)
                    print(begining,end,speeds[begining:end],' == ',speeds[indices[i]:indices[j]])"""
                else:
                    break
                    cycle=False
                    """print(False)
                    print(begining,end,speeds[begining:end],' != ',speeds[indices[i]:indices[j]])"""

                begining=end
                end+=long_seq_to_check
            cycle and potential_cycles.append(speeds[indices[i]:indices[j]]) # si cycle est false we have a court circuit et cette instruction ne s'execute pas
    if (len(potential_cycles) == 0):
        return (False,'0')
    # we search the most compact cycle
    min_cyc=potential_cycles[0]
    min_=len(min_cyc)
    for c in potential_cycles:
        if len(c)<min_:
            min_=len(c)
            min_cyc=c
    return(True,'Tortue cyclique, le tableau des vitesses dans la fenêtre est : '+str(min_cyc))
```

### Identification de la catégorie

Aprés avoir défini les fonctions pour chaque catégorie de tortue, ici on fait appel 
à ces fonctions dans la fonction identify_turtle pour nous afficher le résultat ( la prédiction)
pour chaque tortue.

```python
def indentify_turtle(turtle):
    speeds=list(turtle.diff()[1:]) # we differtiate position to get speed
    reg,msg=regular(speeds)
    if reg :
        return(reg,msg)
    cyc,msg=cyclic(speeds)
    if cyc :
        return(cyc,msg)
    
    
    dis,msg=distraite(speeds)
    return(dis,msg)
```
